```python
import pandas as pd 
data=pd.read_csv('/Users/deen/Desktop/big Data Analytics and Systems/crime_2019_2024.csv')
data.head()
```

```python
crime_data=data.copy()
crime_data=crime_data.drop(['X Coordinate','Y Coordinate','Updated On'], axis=1, errors='ignore')
crime_data=crime_data.dropna()
crime_data
```

```python
crime_data.isna().sum()
```

```python
# Convert all to proper datetime
crime_data['Date'] = pd.to_datetime(crime_data['Date'], errors='coerce', infer_datetime_format=True)

# Drop rows where date couldn't be parsed
crime_data =crime_data.dropna(subset=['Date'])
print("Rows after cleaning:", len(crime_data))

#handling boolean datatype
crime_data['Arrest'] = crime_data['Arrest'].astype(str).str.lower().map({'true': 1, 'false': 0})
crime_data['Domestic'] = crime_data['Domestic'].astype(str).str.lower().map({'true': 1, 'false': 0})
crime_data.to_csv("chicago_crime_2019_2024_clean.csv", index=False)
crime_data.info()
```
